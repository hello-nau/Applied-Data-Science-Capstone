## Part 1 (week 3 Assignment) : Data from Wikipedia page in pandas dataframe (neighborhoods in Toronto)

```python
#import libraries
from lxml.html import fromstring
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0' }
```

```python
#extracting the info
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
hd = requests.get(url, headers = headers)
```

```python
#checking if it was connected properly
hd.status_code
```

```python
soup = BeautifulSoup(hd.content, 'html.parser' )
```

```python
table = soup.find_all('table', class_= 'wikitable sortable')
```

```python
#checking if there is only one table on (if we found everything correctly)
len(table)
```

```python
#extracting the only element lacated in the table
table = table[0]
```

```python
#extract the content of a table
content = [content.get_text() for content in table.find_all('td')]
content
```

```python
#converting data into pd
head = ['Postal code', 'Borough', 'Neighborhood']
data = list(zip(*[iter(content)]*3))
t_n = pd.DataFrame(data[0:], columns = head)
t_n['Neighborhood'] = t_n['Neighborhood'].str.rstrip('\n')
t_n.head(10)
```

```python
#remove that annoying \n
t_n['Borough'] = t_n['Borough'].str.rstrip('\n')
t_n['Postal code'] = t_n['Postal code'].str.rstrip('\n')
t_n.head(10)
```

```python
# count values to see how many are not assigned
t_n['Borough'].value_counts()
```

```python
#replace "Not assigned" to NaN
t_n.replace("Not assigned", np.nan, inplace = True)
t_n.head(10)
```

```python
#replace empty rows to NaN
t_n.replace("", np.nan, inplace = True)
t_n.head(10)
```

```python
#drop NaN cells 
t_n.dropna(axis=0, inplace = True)
```

```python
#rename 'Postal code' to 'Postal Code' so we can merge two dataframes later 
t_n.rename(columns={'Postal code':'Postal Code'},inplace=True)
```

```python
#print the number of rows of a data frame
t_n.shape
```

## Part 2 (week 3 Assignment) : adding latitude and longitude columns

```python
#read the csv file from the url 
path = 'https://cocl.us/Geospatial_data'
pc = pd.read_csv(path)
pc.head(5)
```

```python
#merging two dataframes into one
df_final = pd.merge(t_n, pc, on = 'Postal Code')
```
